```python
# independent and dependent variable

X=df.drop(['child_per_women','Year'],axis=1)
y=df['child_per_women']
```

```python
X.corr()
```

```python
from scipy import stats


for i in X.columns:
    plt.figure(figsize=(14,6))
    plt.subplot(121)
    sns.histplot(X[i])
    plt.title(i)
    
    plt.subplot(122)
    stats.probplot(X[i],dist='norm',plot=plt)
    plt.title(i)
    
    plt.show()
```

```python
# boxcox transformation

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()

X_boxcox=pt.fit_transform(X)

X_boxcox=pd.DataFrame(X_boxcox,columns=X.columns)

for i in X_boxcox.columns:
    
    plt.figure(figsize=(14,7))
    plt.subplot(121)
    sns.histplot(X_boxcox[i])
    plt.title(i)
    
    plt.subplot(122)
    stats.probplot(X_boxcox[i],dist='norm',plot=plt)
    plt.title(i)
    
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
l_reg=LinearRegression()

l_reg.fit(X,y)
```
