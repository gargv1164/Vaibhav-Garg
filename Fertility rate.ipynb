```python
# independent and dependent variable

X=df.drop(['child_per_women','Year'],axis=1)
y=df['child_per_women']
```

```python
#plotting heat map
plt.figure(figsize=(8,8))
sns.heatmap(X.corr(),annot=True)
plt.show()
```

```python
from scipy import stats


for i in X.columns:
    plt.figure(figsize=(14,6))
    plt.subplot(121)
    sns.histplot(X[i])
    plt.title(i)
    
    plt.subplot(122)
    stats.probplot(X[i],dist='norm',plot=plt)
    plt.title(i)
    
    plt.show()
```

```python
# boxcox transformation

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()

X_boxcox=pt.fit_transform(X)

X_boxcox=pd.DataFrame(X_boxcox,columns=X.columns)

for i in X_boxcox.columns:
    
    plt.figure(figsize=(14,7))
    plt.subplot(121)
    sns.histplot(X_boxcox[i])
    plt.title(i)
    
    plt.subplot(122)
    stats.probplot(X_boxcox[i],dist='norm',plot=plt)
    plt.title(i)
    
    plt.show()
```

```python

X_train=X_boxcox.loc[0:14,:]
y_train=df.loc[0:14,:]['child_per_women']


X_test=X_boxcox.loc[15:18,:]
y_test=df.loc[15:18,:]['child_per_women']
```

```python
# ordinary least square estimation 

import statsmodels.api as sm


X_ols=sm.add_constant(X_train)
model=sm.OLS(y_train,X_ols).fit()

model.summary()
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso_reg=Lasso()

params = {"alpha" : [1e-3, 1e-2, 1, 1e1, 
                     1e2],
         'tol':[0.05,0.1,0.15],
         'max_iter':[50,100,200]}

grid=GridSearchCV(lasso_reg,params,scoring='neg_mean_squared_error',cv=10)

grid.fit(X_train,y_train)
```

```python
print(grid.best_score_,
      grid.best_estimator_)
```

```python
# fitting lasso with best estimated parameters

lasso_best=grid.best_estimator_
lasso_best.fit(X_train,y_train)
```

```python
lasso_best.coef_
```

```python
from sklearn.linear_model import LinearRegression
l_reg=LinearRegression()

l_reg.fit(X_train,y_train)
```

```python
predicted_y=l_reg.predict(X_test)
print(predicted_y)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,predicted_y)
```

```python
# principle component analysis

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(X_boxcox)
X_pca=pca.transform(X_boxcox)
```

```python
print('Dimensions before reduction : {} \n\nDimensions after reduction : {}'.format(X_boxcox.shape, X_pca.shape))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.show()
```

```python
# predictions after PCA


X_pca_train=X_boxcox.loc[0:14,:]
X_pca_test=X_boxcox.iloc[[15,16,17,18],:]
y_pca_train=df.loc[0:14,:]['child_per_women']
y_pca_test=df.iloc[[15,16,17,18],1]


from sklearn.linear_model import LinearRegression
l_reg_pca=LinearRegression()

l_reg_pca.fit(X_pca_train,y_pca_train)

predicted_y_pca=l_reg_pca.predict(X_pca_test)
print(predicted_y_pca)
```

```python
mean_squared_error(y_pca_test,predicted_y_pca)
```
