```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('fertility rate in india.xlsx')
data.head(10)
```

```python

plt.figure(figsize=(15,8))
plt.scatter(data['Year'],data['Fertility Rate'])
plt.xlabel('Years')
plt.ylabel('Fertility Rate')
```

```python
data2=pd.read_excel('fertility by state.xlsx')
data2.head()
```

```python
plt.figure(figsize=(20,8))

plt.plot(data2['State/UT'],data2['Fertility rate 2021'],label='2021')

plt.plot(data2['State/UT'],data2['Fertility rate 1981[2]'],label='1981',)
plt.plot(data2['State/UT'],data2['fertility_2009'],label='2009')
plt.plot(data2['State/UT'],data2['Fertility rate 1999[6]'],label='1999')
plt.plot(data2['State/UT'],data2['Fertility rate 1991[2]'],label='1991')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
